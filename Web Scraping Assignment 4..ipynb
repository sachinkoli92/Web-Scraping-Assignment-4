# Q2. Scrape the details of team India's Interantional fixtures from bcci.tv.

```python
#Lets first connect to web driver
driver=webdriver.Chrome("C:\web driver\chromedriver.exe")
```

```python
driver.get('https://www.bcci.tv/')
driver.maximize_window()
```

```python
# Clicking on international button.
driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover'][1]").click()
```

```python
# Clicking on fixture button.
fixture=driver.find_element_by_xpath("//div[@class='drop-down__options']//li[1]//a").click()
```

```python
#Creating empty lists to store scrape data.

fixtures={}
fixtures['Match_title']=[]
fixtures['Series']=[]
fixtures['Place']=[]
fixtures['Date/Month']=[]
fixtures['Time']=[]

#extracting Match title

try:
    for i in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/strong'):
        fixtures['Match_title'].append(i.text)
except NoSuchElementExceptions:
    fixtures['Match_title'].append('--')
    
#extracting Series

try:
    for i in driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]'):
        fixtures['Series'].append(i.text)
except NoSuchElementExceptions:
    fixtures['Series'].append('--')
    
#extracting Place

try:
    for i in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span'):
        fixtures['Place'].append(i.text)
except NoSuchElementExceptions:
    fixtures['Place'].append('--')
    
#extracting Date and month as they both are having different tags
date=[]
month=[]
try:
    for i in driver.find_elements_by_xpath('//div[@class="fixture__full-date"]/span'):
        date.append(i.text)
except NoSuchElementException:
    date.append('--')
    
try:
    for i in driver.find_elements_by_xpath('//div[@class="fixture__date-details"]/span[1]'):
        month.append(i.text)
except NoSuchElementException:
    month.append('--')
    
#Combining date and month
for i in range(len(month)):
    fixtures['Date/Month'].append(date[i]+month[i])
    
#extracting time

try:
    for i in driver.find_elements_by_xpath('//div[@class="fixture__date-details"]/span[2]'):
        fixtures['Time'].append(i.text)
except NoSuchElementExceptions:
    fixtures['Time'].append('--')
        
```

```python
fixture=pd.DataFrame.from_dict(fixtures)
fixture
```

```python
driver.get('https://archive.ics.uci.edu/ml/index.php')
driver.maximize_window()
```

```python
# Clicking on the view all data sets.
driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/span/b/a').click()
```

```python
#Creating empty lists for the information of UCI Datasets.
dataset={}
dataset['name']=[]
dataset['datatype']=[]
dataset['task']=[]
dataset['attributetype']=[]
dataset['no of instances']=[]
dataset['no of attributes']=[]
dataset['year']=[]

# extracting dataset name

try:
    for i in driver.find_elements_by_xpath("//p[@class='normal']//b/a"):
        dataset['name'].append(i.text)
except NoSuchElementException:
    dataset['Name'].append('--')

for i in range (2,587):
    
# extracting data type

    try:
        dataset['datatype'].append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[2]/p'.format(i)).text)
    except NoSuchElementException:
        dataset['datatype'].append('--')
        
#extracting task

    try:
        dataset['task'].append(driver.find_element_by_xpath('//table[@border="1"]//tr[{}]//td[3]'.format(i)).text)
    except NoSuchElementException:
        dataset['task'].append('--')
        
#extracting attribute types

    try:
        dataset['attributetype'].append(driver.find_element_by_xpath('//table[@border="1"]//tr[{}]//td[4]'.format(i)).text)
    except NoSuchElementException:
        dataset['attributetype'].append('--')
        
#extracting Number of instances

    try:
        dataset['no of instances'].append(driver.find_element_by_xpath('//table[@border="1"]//tr[{}]//td[5]'.format(i)).text)
    except NoSuchElementException:
        dataset['no of instances'].append('--')
        
#extracting Number of attributes

    try:
        dataset['no of attributes'].append(driver.find_element_by_xpath('//table[@border="1"]//tr[{}]//td[6]'.format(i)).text)
    except NoSuchElementException:
        dataset['no of attributes'].append('--')
        
#extracting Year

    try:
        dataset['year'].append(driver.find_element_by_xpath('//table[@border="1"]//tr[{}]//td[7]'.format(i)).text)
    except NoSuchElementException:
        dataset['year'].append('--')
```
